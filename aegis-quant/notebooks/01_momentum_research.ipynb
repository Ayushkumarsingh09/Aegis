# Momentum Strategy Research

Demonstrates the full Aegis Quant research pipeline.

```python
from aegis_quant.data.engine import MarketDataEngine
from aegis_quant.data.connectors import CSVConnector
from aegis_quant.features.engine import FeatureEngine
from aegis_quant.strategy.strategies import MomentumStrategy, run_strategy
from aegis_quant.risk.metrics import RiskEngine

engine = MarketDataEngine()
engine.ingest(CSVConnector('data/sample'), 'BTC-USD')
df = engine.get_bars('BTC-USD')
df = FeatureEngine().compute_all(df)
result = run_strategy(MomentumStrategy('BTC-USD'), df)
print('Metrics:', result.metrics)
print('Trades:', len(result.fills))
```
